# Load the dataset

```python
import pandas as pd

bike_rentals = pd.read_csv("Bike-Sharing-Dataset/hour.csv")
bike_rentals.head()
```

# Analyze the dataset

Display the distribution for the 'cnt' column

```python
import matplotlib.pyplot as plt

bike_rentals.cnt.hist()
plt.show()
```

Display the correlations between the features

```python
bike_rentals.corr()
```

# Create additional features

Add `time_label` feature which categorizes the hour of the day in values representing morning, afternoon, evening, and night

`hour` | `time_label`  
--- | --- 
0 | night
1 | morning
2 | afternoon
3 | evening

```python
def assign_label(hour):
    if hour <= 6:
        return 4
    elif hour <= 12:
        return 1
    elif hour <= 18:
        return 2
    elif hour <= 24:
        return 3

bike_rentals["time_label"] = bike_rentals.hr.apply(assign_label)
bike_rentals.head(10)
```

# Split into train and test sets

```python
train = bike_rentals.sample(frac=0.8)
test = bike_rentals.loc[~bike_rentals.index.isin(train.index)]

print(train.shape[0])
print(test.shape[0])
print(bike_rentals.shape[0])
```

## Select predictors

```python
import numpy as np

cnt_correlations = bike_rentals.corr()["cnt"]
cnt_correlations = np.abs(cnt_correlations)
cnt_correlations = cnt_correlations.sort_values(ascending=False)
print(cnt_correlations)
```

```python
cnt_correlations = cnt_correlations[3:]
print(cnt_correlations)
```

```python
cnt_correlations = cnt_correlations[cnt_correlations > 0.1]
predictors = cnt_correlations.index.tolist()
print(predictors)
```

## Fit a linear regression and calculate the MSE

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lr = LinearRegression()

lr.fit(train[predictors].values, train.cnt)
predictions = lr.predict(test[predictors].values)
mse = mean_squared_error(test.cnt, predictions)
print(mse)
```

## Fit a decision tree and calculate the MSE

```python
from sklearn.tree import DecisionTreeRegressor

clf = DecisionTreeRegressor(random_state=1)
clf.fit(train[predictors], train.cnt)

test_predictions = clf.predict(test[predictors])
mse = mean_squared_error(test.cnt, test_predictions)
print(mse)
```

## Fit a random forest and calculate the MSE

```python
from sklearn.ensemble import RandomForestRegressor

clf = RandomForestRegressor(n_estimators=5)
clf.fit(train[predictors], train.cnt)

test_predictions = clf.predict(test[predictors])
mse = mean_squared_error(test.cnt, test_predictions)
print(mse)
```
